#Análise de Sentimentos em base com rótulos de 5 emoções

```
# Importação das bibliotecas necessárias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import keras 
import sklearn.model_selection as sk
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.optimizers import SGD
from keras.optimizers import Adam
```

Upload do arquivo considerando a vírgula como separador

```
imagem_face = pd.read_csv('imagedb_CH_disciplina.csv',sep=',', encoding='UTF-8', header=None)
imagem_face.head(50)
```

# Cleaning na Base de Dados
Gerar arquivo apenas com as entradas que representam as emoções FEAR ou ANGER ou JOY ou SADNESS

```
imagem_face[:][1]=='admiration'
imagem_face.loc[imagem_face[:][1]=='admiration']
base_treino = imagem_face.loc[(imagem_face[:][1]== 'fear') | (imagem_face[:][1]=='anger') | (imagem_face[:][1]=='joy') | (imagem_face[:][1]=='sadness') | (imagem_face[:][1]=='neutral')]
contador_base_treino = len(base_treino)
```

```
contador_base_treino
```

```
base_treino.head(50)
```

Geração de arquivo cujo conteúdo são os pares de coordenadas de pontos característicos da face das entradas com as emoções descritas acima.

```
entrada = base_treino.iloc[0:base_treino.shape[0],9:base_treino.shape[1]]
entrada.head()
```

```
rotulo = base_treino.iloc[0:base_treino.shape[0],1]
rotulo.head()
```

Organização da base com o objetivo de tornar os pares das coordenadas campos do Data Frame a ser gerado para treino e validação

```
dados = np.zeros(shape=(entrada.shape[0],2*entrada.shape[1]))
for j in range(entrada.shape[1]):
  for i in range(entrada.shape[0]):
    par = entrada.iloc[i,j].split(';')
    dados[i,2*j] = float(par[0].split('(')[1])
    dados[i,2*j+1] = float(par[1].split(')')[0])
input = pd.DataFrame(data=dados)
```

Anotação dos rótulos no Data Frame a ser gerado para treino e validação

```

emotion = np.zeros(shape=(rotulo.shape[0],5))
for i in range(rotulo.shape[0]):
    if rotulo.iloc[i] == 'fear':
      emotion[i][0] = 1
    elif rotulo.iloc[i] == 'anger':
      emotion[i][1] = 1
    elif rotulo.iloc[i] == 'joy':
      emotion[i][2] = 1
    elif rotulo.iloc[i] == 'sadness':
      emotion[i][3] = 1
    elif rotulo.iloc[i] == 'neutral':
      emotion[i][4] = 1
classe_emotion = pd.DataFrame(data=emotion)
classe_emotion.head(20)
```

# Aplicação do Modelo de Classificador
Rede Neural com 3 camadas e uso da função de Ativação RELU e SoftMax na útlima camada para identificação da probabilidade de pertencer a cada uma das classes

```
X_treina, X_testa, Y_treina, Y_testa = sk.train_test_split(dados, emotion, test_size=0.2, random_state=9)
```

```
model = Sequential()
model.add(Dense(80, activation='tanh', input_dim=112))
# model.add(Dropout(0.5))
model.add(Dense(80, activation='tanh'))
# model.add(Dropout(0.5))
model.add(Dense(5, activation='softmax'))

adam = Adam(lr=0.001)
# sgd = SGD(lr=0.01, decay=1e-6)
model.compile(loss='mean_squared_error',
              optimizer=adam,
              metrics=['accuracy'])

model.fit(X_treina, Y_treina,
          epochs=100)
score = model.evaluate(X_testa, Y_testa)
```
